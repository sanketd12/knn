````python
from keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np
df = mnist.load_data('minst.db')
train,test = df
X_train, y_train = train
X_test, y_test = test

np.set_printoptions(edgeitems=30, linewidth=100000)
for i in range(5): 
    print(y_test[i],'\n') # Print the label
    print(X_test[i],'\n') # Print the matrix of values
    plt.contourf(np.rot90(X_test[i].transpose())) # Make a contour plot of the matrix values
    plt.show()


````

X_train has 60,000 images, and X_test has 10,000 images, with each image being 28×28 pixels. A single image, X_train[i] or X_test[i], has a shape of (28, 28). The labels, y_train and y_test, have 60,000 and 10,000 values, respectively, with each representing a digit from 0 to 9.

````python
# Question 3
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from sklearn.neighbors import KNeighborsClassifier

(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train.reshape(X_train.shape[0], 784)
X_test = X_test.reshape(X_test.shape[0], 784)

print("Reshaped X_train shape:", X_train.shape)
print("Reshaped X_test shape:", X_test.shape)

#Question 4
# Determine optimal k:
k_bar = 20
accuracy = []

for k in range(1, k_bar + 1):
    print('Current k: ', k)
    knn = KNeighborsClassifier(n_neighbors=k)  # Create classifier instance
    predictor = knn.fit(X_train, y_train)  # Fit model
    accuracy.append(predictor.score(X_test, y_test))  # Store accuracy

# Find the optimal k
accuracy_max = np.max(accuracy)  # highest recorded accuracy
max_index = np.argmax(accuracy)  # index of highest accuracy
k_star = max_index + 1  # Optimal k is index+1 since k starts from 1
print("Optimal k:", k_star)

# Plot accuracy by k
plt.plot(range(1, k_bar + 1), accuracy, marker='o')
plt.xlabel("k")
plt.ylabel("Accuracy")
plt.title("Optimal k: " + str(k_star))
plt.grid()
plt.show()
````

````python
# Question 5
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns

# Train model with optimal k
optimal_knn = KNeighborsClassifier(n_neighbors=k_star)
optimal_knn.fit(X_train, y_train)
y_pred = optimal_knn.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Test Set Accuracy with k={k_star}: {accuracy:.4f}")

# Compute confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Plot matrix
plt.figure(figsize=(8,6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=range(10), yticklabels=range(10))
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title(f"Confusion Matrix (k={k_star})")
plt.show()
````

**Question 7 Response:**
To represent a color photo in matrix form, we can break the image down into a grid of pixels. Each pixel is represented by three values corresponding to its RGB (Red, Green, Blue) color channels. For example, a 100x100 pixel image would be represented as a 100x100x3 matrix, where each element is a color value from 0 to 255. To convert this matrix into tabular data, you would need to flatten the entire matrix into a one-dimensional vector, thus treating each pixel's RGB values as individual features in the dataset. This allows for you to apply ML techniques to analyze and classify the image just like other tabular data.
